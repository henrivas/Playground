```python
def subtract_all(*numbers):
    result = numbers[0]
    for num in numbers[1:]:
        result -= num
    return result
subtract_all(10, 5, 1)
```

```python
import re

def no_spaces(legalname):
    spaced = re.sub(r'([a-z])([A-Z])', r'\1 \2', legalname)
    return spaced

invoice_value = "NasdaqStockholmAB"
output = no_spaces(invoice_value)
print(output)
```

```python
def add_spaces(field_value):
    result = []
    for w in range(len(field_value)):
        if w > 0 and field_value[w].isupper() and field_value[w - 1].islower():
            result.append(' ')
        result.append(field_value[w])
    return ''.join(result)

field_value = "NasdaqTechStockholmLTD"
field_value = add_spaces(field_value)
print(field_value)
```
