# Sukurti savo klasÄ™, kuri turÄ—tÅ³ bent 2 argumentus __init__ metodui bei bent 3 papildomus metodus.

```python
class bankomatas:
    def __init__(self, sask_nr, prad_likutis):
        self.sask_nr = sask_nr
        self.balansas = prad_likutis

    def isidejimas(self, kiekis):
        if kiekis >= 0:
            self.balansas += kiekis
            return f"Isidejote {kiekis} euru. Saskaitos likutis: {self.balansas} eurai"

    def issiemimas(self, kiekis):
        if kiekis > 0 and kiekis <= self.balansas:
            self.balansas -= kiekis
            return f"Issiemete {kiekis} euru. Saskaitos likutis: {self.balansas} eurai"

    def likutis(self):
        return f"Saskaitos *{self.sask_nr}* likutis: {self.balansas} eurai"

bankas = bankomatas("DE98237547823", 1000)
print(bankas.likutis())
print(bankas.isidejimas(500))
print(bankas.issiemimas(300))
```
