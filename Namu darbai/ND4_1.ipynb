```python
nuo = int(input('Iveskite pirmaji skaiciu'))
iki = int(input('Iveskite antraji skaiciu'))
step = int(input('Iveskite treciaji skaiciu'))
def skaiciuKurejas(nuo, iki, step):
    result = []
    current = nuo
    while current < iki:
        result.append(current)
        current += step
    return result
rezultatas = skaiciuKurejas(nuo, iki, step)
print(rezultatas)
```
