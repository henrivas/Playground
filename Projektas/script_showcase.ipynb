```python
field_value1 = "1 234,56"

import re
if " " in field_value1:
  field_value1 = field_value1.replace(" ", "")
if "," in field_value1:
  field_value1 = field_value1.replace(",", ".")
print(field_value1)
```

```python
field_value2 = "PO#73699"

x = re.findall("[0-9.]+", field_value2)
 
field_value2 = x[0]
print(field_value2)
```

```python
field_value3 = "ProgramavimoMokyklaUAB"

def add_spaces(field_value3):
  result = []
  for w in range(len(field_value3)):
    if w > 0 and field_value3[w].isupper() and field_value3[w - 1].islower():
      result.append(' ')
    result.append(field_value3[w])
  return ''.join(result)
field_value3 = add_spaces(field_value3)

print(field_value3)
```

```python
field_value4 = "Moms 25% 20 995,00 (64 874,00)"

x = re.findall(r'\b(\d+(\,\d{3})*(\.\d+)?)(?=%)', field_value4)

field_value4 = x[0][0]
print(field_value4)
```

```python
field_value5 = "Programavimo Mokykla Sweden Aktiebolag"

x = field_value5.replace('Aktiebolag', 'AB')

field_value5 = x
print(field_value5)
```

```python
import pandas as pd

table_values = {
    "Item_Description": None,
    "Quantity": None,
    "Item_Total": '10000',
    "Item_Tax_Amount": None,
    "Item_Tax": '25',
}

table_values_df = pd.DataFrame.from_dict(table_values , orient='index').T
table_values_df.columns = ["Item_Total", "Item_Tax_Amount", "Item_Tax", "Quantity", "Item_Description"]
table_values_df.index.name = "0"

import pandas as pd
import re
 
df = pd.DataFrame.from_dict(table_values , orient='index')
df["Item_Total"] = df["Item_Total"].replace({',':'.'},regex=True).replace({' ':''},regex=True)
df["Item_Tax_Amount"] = df["Item_Tax_Amount"].replace({',':'.'},regex=True).replace({' ':''},regex=True)
df["Item_Tax"] = df["Item_Tax"].replace({',':'.'},regex=True).replace({' ':''},regex=True)
df["Quantity"] = df["Quantity"].replace({',':'.'},regex=True).replace({' ':''},regex=True)
df["Item_Tax"] = df["Item_Tax"].replace({'%':''},regex=True)
df["Item_Total"] = df["Item_Total"].astype(float) / (df["Item_Tax"].astype(int)/100 + 1)
df["Item_Tax_Amount"] = (df["Item_Total"].astype(float) * (df["Item_Tax"].astype(int)/100 + 1)) - df["Item_Total"].astype(float)

table_values = df.to_dict()
print(table_values_df)
```
